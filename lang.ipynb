```python
#create sqlalchemy/ sqldb obj
from langchain_community.utilities import SQLDatabase
import os 

pg_uri =  os.getenv("pg_uri") or os.getenv("pg_uri".upper()) or globals().get("pg_uri")
if not pg_uri:
    raise ValueError("pg_uri is not defined.Make sure u set pg_uri earlier.")
db = SQLDatabase.from_uri(pg_uri)
print(db.get_usable_table_names())
```

```python
#creation of sql chain 
#using sqldb [llm + given db] 
# translates questions -> SQL -> run them
from langchain_community.utilities import SQLDatabase
from langchain_community.agent_toolkits.sql.base import create_sql_agent
#from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq

# return_direct=False ensures chain gives a composed natural-language answer
sql_chain = create_sql_agent(
    llm=llm,        #llm gen SQL n answers
    db=db,
    verbose=True,           # prints SQL executed in notebook output (good for debugging)
    return_direct=False     # return natural-language response, not raw SQL only
)
```

```python
#safety: restrict destructive operations
"""By default the chain will not intentionally issue DROP/ALTER etc, but it's 
best to proactively block dangerous queries. Hereâ€™s a small wrapper to check the model-generated 
SQL and deny/raise if dangerous keywords are present."""
from sqlalchemy import text
DANGEROUS = {"drop", "delete", "truncate", "alter", "create", "shutdown"}

def is_safe_sql(sql: str) -> bool:
    s = sql.lower()
    return not any(token in s for token in DANGEROUS)

def safe_run(query: str):
    """Run natural language query via sql_chain but block dangerous SQL."""
    # ask the chain to produce SQL but because SQLDatabaseChain executes internally,
    # we'll rely on verbose=True printing the SQL. If you want the SQL separately,
    # use chain.tools or implement a custom agent pattern.
    out = sql_chain.run(query)
    return out
```

```python
#create chatbot function
#NLP -> SQL -> PostgreSQL -> Final Answer
def chatbot(question: str):
    print("User:", question)
    
    try:
        #Run SQL Agent
        response = sql_chain.run(question)
        print("Chatbot Answer:")
        return response
    except Exception as e:
        print("Error:", str(e))
        return "I faced an error while answering ur question"
```

```python
#connecting notebook -> backend -> langsmith
import requests 

#BACKEND_URL = "http://127.0.0.1:5000/ask"
url = "https://nlp2sql-chatbot.onrender.com/ask"


def ask_backend(question: str):
    payload = {"question": question}
    response = requests.post(url, json=payload)#BACKEND_URL, json=payload)

    if response.status_code == 200:
        return response.json()
    else:
        print("Error:", response.text)


ask_backend("Show me all customers")
```
